```python
movie_mat = df.pivot_table(index = 'user_id', columns= 'title', values = 'rating')
```

```python
movie_mat  #NaN means he has not watched the movie. i.e., he has not given any ratings
```

```python
# To find similarities between the below mentioned two movies, take any two movies and try to find similarities

starwars_rating = movie_mat['Star Wars (1977)']
angry_men_rating = movie_mat['12 Angry Men (1957)']
```

```python
similar_to_starwars = movie_mat.corrwith(starwars_rating)
```

```python
similar_to_starwars
```

```python
similar_to_starwars.sort_values(ascending=False).head(10)  # correlation = 1 , means perfect correction
```

```python
similar_angry = movie_mat.corrwith(angry_men_rating)
```

```python
similar_angry.sort_values(ascending=False).head(10)
```
